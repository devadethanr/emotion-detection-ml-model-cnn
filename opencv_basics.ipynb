```python
import cv2
from matplotlib import pyplot as plt
```

```python
# plotting images
img_1 = cv2.imread("resource/lena1.jpeg")
plt.imshow(img_1)
```

```python
img_2 = cv2.imread("resource/baboon.jpg")
plt.imshow(img_2)
```

```python
#ADDDITION
result_plt = img_1+img_2
plt.imshow(result_plt)
```

```python
#SUBSTRACTION
result_plt = img_1-img_2
plt.imshow(result_plt)
```

```python
#mul
result_plt = img_1*img_2
plt.imshow(result_plt)
```

```python
#division
result_plt = img_2/img_1
plt.imshow(result_plt)
```

blur and other

```python
#blur
img_3 =  cv2.imread("resource/brain.png")
img = cv2.cvtColor(img_3,cv2.COLOR_BGR2HSV)
figure_size = 7
result_blur = cv2.blur(img_3,(figure_size,figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(cv2.cvtColor(img,cv2.COLOR_HSV2RGB)),plt.title("original")
plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.subplot(122),plt.imshow(cv2.cvtColor(result_blur,cv2.COLOR_HSV2RGB)),plt.title("blurred")
plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.show()
```

```python
#resize and zooming
img_1 = cv2.imread("resource/lena1.jpeg")
zoom_img = cv2.resize(img_1,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom_img)
```

```python

height, width, channels = img_1.shape # use this to define height ,width and channels 
zoom_img = cv2.resize(img_1,(int(width*0.4),int(height*0.4)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom_img)
```

```python
#rotation
rotationmet = cv2.getRotationMatrix2D((width/2,height/2),30,1) #for rotation matrix
rotate = cv2.warpAffine(img_1,rotationmet,(height,width)) #for wrapping image
plt.imshow(rotate)
```

```python
import numpy as np


trans_val = np.float32([[1,0,50],[0,1,100]])
dist = cv2.warpAffine(img_1,trans_val,(height,width))
plt.imshow(dist)
```
